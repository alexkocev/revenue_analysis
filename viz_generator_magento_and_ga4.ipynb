```python
# -- CORRELATION ANALYSIS --
import pandas as pd
import numpy as np
from scipy.stats import pearsonr

# Load the data from the CSV file
file_path = "data/export_M2.csv"  # Replace with your actual file path
data = pd.read_csv(file_path)

# Ensure numeric data
data['Grand Total (Base)'] = pd.to_numeric(data['Grand Total (Base)'], errors='coerce')

# Calculate the first purchase value for each customer
data['First Purchase Value'] = data.groupby('Customer Name')['Grand Total (Base)'].transform('first')

# Group by customer to calculate total and first purchase revenues
client_data = data.groupby('Customer Name').agg({
    'Grand Total (Base)': 'sum',  # Lifetime Value (LTV)
    'First Purchase Value': 'first'  # First Purchase Value (FPV)
}).rename(columns={'Grand Total (Base)': 'LTV', 'First Purchase Value': 'FPV'})

# Calculate the correlation
correlation, p_value = pearsonr(client_data['FPV'], client_data['LTV'])

# Find illustrative examples
high_ltv_examples = client_data.sort_values('LTV', ascending=False).head(5)

# Output results
print(f"Correlation between FPV and LTV: r = {correlation:.2f}")
print(f"P-value of correlation: {p_value:.5f}")
print("\nExamples of high LTV clients:")
print(high_ltv_examples)

# Additional Analysis: Ratio of FPV to LTV
client_data['FPV_to_LTV_Ratio'] = (client_data['FPV'] / client_data['LTV']) * 100
low_ratio_clients = client_data[client_data['FPV_to_LTV_Ratio'] < 10].sort_values('LTV', ascending=False).head(5)

print("\nHigh LTV clients with FPV less than 10% of LTV:")
low_ratio_clients
```

```python
# -- FIRST PURCHASE VS. TOTAL REVENUE --
import pandas as pd
import plotly.graph_objects as go

# Load the data from the CSV file
file_path = "data/export_M2.csv"  # Replace with your actual file path
data = pd.read_csv(file_path)

# Ensure numeric data
data['Grand Total (Base)'] = pd.to_numeric(data['Grand Total (Base)'], errors='coerce')

# Calculate the first purchase value for each customer
data['First Purchase Value'] = data.groupby('Customer Name')['Grand Total (Base)'].transform('first')

# Group by customer to calculate total and first purchase revenues
client_data = data.groupby('Customer Name').agg({
    'Grand Total (Base)': 'sum',
    'First Purchase Value': 'first'
})

# Calculate metrics
total_revenue = client_data['Grand Total (Base)'].sum()
first_purchase_revenue = client_data['First Purchase Value'].sum()
revenue_less_first_purchase = total_revenue - first_purchase_revenue

# Create a Plotly bar chart
fig = go.Figure()

# Add bars for first purchase revenue and remaining revenue
fig.add_trace(go.Bar(
    x=["First Purchase Revenue", "All Revenue less First Purchase"],
    y=[first_purchase_revenue / total_revenue * 100, revenue_less_first_purchase / total_revenue * 100],
    text=[f"{first_purchase_revenue / total_revenue * 100:.2f}%", f"{revenue_less_first_purchase / total_revenue * 100:.2f}%"],
    textposition='auto',
    marker=dict(color=['blue', 'skyblue'])
))

# Customize the layout
fig.update_layout(
    title="First Purchase vs. Total Revenue",
    xaxis_title="Revenue Breakdown",
    yaxis_title="Percentage of Total Revenue",
    yaxis=dict(ticksuffix="%"),
    template="plotly_white",
    width=700,
    height=500
)

# Show the chart
fig.show()
```

```python
# -- CLIENT ORIGIN --
import pandas as pd
import numpy as np

# Load the data from the CSV file
file_path = "data/client_origin.csv"  # Replace with your actual file path
df = pd.read_csv(file_path)




# Filter top 10 based on Item revenue
df = df.sort_values(by="Item revenue", ascending=False).head(10)

df["User Origin"] = df["First user source"]


# Calculate additional metrics
total_revenue = df["Item revenue"].sum()
total_items = df["Items purchased"].sum()

df["% of Revenue"] = (df["Item revenue"] / total_revenue) * 100
df["% of Items Purchased"] = (df["Items purchased"] / total_items) * 100
df["Average Revenue per Item"] = df["Item revenue"] / df["Items purchased"]

# Replace NaN or inf values (e.g., for rows with 0 items purchased)
df["Average Revenue per Item"] = df["Average Revenue per Item"].fillna(0).replace([float('inf'), -float('inf')], 0)

# Format percentages for readability
df["% of Revenue"] = df["% of Revenue"].map("{:.2f}%".format)
df["% of Items Purchased"] = df["% of Items Purchased"].map("{:.2f}%".format)
df["Average Revenue per Item"] = df["Average Revenue per Item"].map("${:,.2f}".format)

# Select and rename columns for the final table
client_origin_table = df[[
    "User Origin",
    "% of Revenue",
    "% of Items Purchased",
    "Average Revenue per Item"
]].rename(columns={
    "First user source": "User Origin",
    "% of Revenue": "% of Revenue",
    "% of Items Purchased": "% of Items Purchased",
    "Average Revenue per Item": "Avg Revenue per Item"
})

# Display the final table
client_origin_table
```
